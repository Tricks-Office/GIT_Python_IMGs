### 명암 기반 이미지 판정 원본 소스

```python
import cv2
import numpy as np

def process_image(image_path):
    # 이미지 로드
    image = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)
    
    if image is None:
        print("이미지를 불러오지 못했습니다.")
        return

    # 이미지 이진화 (기본 임계값 127 사용, 필요시 조정 가능)
    _, binary_image = cv2.threshold(image, 127, 255, cv2.THRESH_BINARY)

    # 검은색 영역 찾기 (명도 값이 0인 부분)
    black_area = image[binary_image == 0]

    # 검은색 영역 내 명도가 10 초과하는지 확인
    if np.any(black_area > 10):
        print("NG")
    else:
        print("OK")

# 이미지 경로 설정
image_path = 'C:\\Data\\Sample\\IMG_Judge\\OK\\Key1.png' # 이미지 경로는 환경에 맞게 수정하세요.

# 이미지 처리
process_image(image_path)
```
