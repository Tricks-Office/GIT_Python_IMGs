```python
import cv2
import numpy as np

# 이미지 로드
input_image_path = r'C:\Data\Sample\IMG\Target\Dog.jpg'
output_image_path_canny = 'canny_edges.jpg'

img = cv2.imread(input_image_path, cv2.IMREAD_GRAYSCALE)

# Canny 엣지 검출
edges = cv2.Canny(img, 100, 200)

# 결과 저장
cv2.imwrite(output_image_path_canny, edges)

# # 결과 출력
# cv2.imshow('Canny Edges', edges)
# cv2.waitKey(0)
# cv2.destroyAllWindows()
import matplotlib.pyplot as plt
plt.imshow(edges)
plt.show()
```

```python
import cv2
import numpy as np

# # 이미지 로드
# input_image_path = 'input_image.jpg'
output_image_path_contours = 'contours.jpg'

img = cv2.imread(input_image_path)
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

# 이진화
ret, thresh = cv2.threshold(gray, 127, 255, cv2.THRESH_BINARY)

# 윤곽선 찾기
contours, hierarchy = cv2.findContours(thresh, cv2.RETR_TREE, cv2.CHAIN_APPROX_SIMPLE)

# 새로운 이미지에 윤곽선 그리기
contour_img = np.zeros_like(img)
cv2.drawContours(contour_img, contours, -1, (255, 255, 255), 1)

# 결과 저장
cv2.imwrite(output_image_path_contours, contour_img)

# # 결과 출력
# cv2.imshow('Contours', contour_img)
# cv2.waitKey(0)
# cv2.destroyAllWindows()

plt.imshow(contour_img)
plt.show()
```

```python
import cv2
import numpy as np
import os

# 폴더 경로 설정
input_folder = r'C:\Data\Sample\IMG\Target'
output_folder_canny = r'C:\Data\Sample\IMG\Canny'
output_folder_contours = r'C:\Data\Sample\IMG\Contours'

for file in os.listdir(input_folder):
    # 이미지 로드
    input_image_path = os.path.join(input_folder,file)
    output_image_path_canny = os.path.join(output_folder_canny,file)
    output_image_path_contours = os.path.join(output_folder_contours, file) 


    img = cv2.imread(input_image_path, cv2.IMREAD_GRAYSCALE)

    # Canny 엣지 검출
    edges = cv2.Canny(img, 100, 200)

    # 결과 저장
    cv2.imwrite(output_image_path_canny, edges)


    img = cv2.imread(input_image_path)
    gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

    # 이진화
    ret, thresh = cv2.threshold(gray, 127, 255, cv2.THRESH_BINARY)

    # 윤곽선 찾기
    contours, hierarchy = cv2.findContours(thresh, cv2.RETR_TREE, cv2.CHAIN_APPROX_SIMPLE)

    # 새로운 이미지에 윤곽선 그리기
    contour_img = np.zeros_like(img)
    cv2.drawContours(contour_img, contours, -1, (255, 255, 255), 1)

    # 결과 저장
    cv2.imwrite(output_image_path_contours, contour_img)
```
