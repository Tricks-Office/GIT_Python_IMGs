```python
import cv2
import numpy as np

# 이미지 경로 설정
image_path = 'C:\\Data\\Sample\\IMG_Judge\\OK\\Key1.png'

# 이미지 처리
# 이미지 로드
image = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)

if image is None:
    print("이미지를 불러오지 못했습니다.")

# 이미지 흐리게 조정
blurred = cv2.blur(image, (5, 5))

# 이미지 이진화 (기본 임계값 127 사용, 필요시 조정 가능)
_, binary_image = cv2.threshold(blurred, 50, 255, cv2.THRESH_BINARY)

# 검은색 영역 찾기 (명도 값이 0인 부분)
black_area = image[binary_image == 0]

# 검은색 영역 내 명도가 10 초과하는지 확인
if np.any(black_area > 10):
    print("NG")
else:
    print("OK")

import matplotlib.pyplot as plt

#이미지 출력
fig, axe = plt.subplots(1, 3, figsize=(10, 5))
axe[0].imshow(image, cmap='gray')
axe[1].imshow(blurred, cmap='gray')
axe[2].imshow(binary_image, cmap='gray') 
plt.show()

cv2.imwrite('Original.png',image)
cv2.imwrite('Binary.png', binary_image)
```
